## HIST PLOT

A histogram is a classical visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within the discrete bins.

This function can normalize the statistic computed within each bin to estimate frequency, density or probability mass, and it can add a smooth curve obtained using a kernel density estimate, similar to # kdeplot()

#### 1. Creating a basic HISTOGRAM on horizontal/vertical axis

```python
sns.histplot(x='Total',data=mart)
```

#### 2. Change the width of Bins in a HISTOGRAM

```python
sns.histplot(x='Total',data=mart,binwidth=120,color='green')
```

#### 3. Changing number of Bins and Interval on X-Axis in a HISTOGRAM PLOT

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total',data=mart,bins=np.arange(0,1100,50))
plt.xticks(np.arange(0,1100,50))
```

#### 4. Including a KDE in a HISTOGRAM PLOT

```python
sns.histplot(x='Total',data=mart,kde=True,color='violet')
```

#### 5. Creating a stack plot using seaborn| Stack the HISTOGRAM PLOT on a categorical variable in hue

```python
plt.figure(figsize=(10,5))
sns.histplot(data=mart,x='Total',bins=np.arange(0,1100,50),hue='Payment',multiple='stack',palette='twilight')
```

#### 6. Creating a step plot using HIST PLOT in seaborn and making it no fill| creating a poly plot

```python
sns.histplot(x='Total',data=mart,element='step',fill=False)
```

```python
sns.histplot(x='Total',data=mart,element='poly',hue='Payment',palette='turbo')
```

#### 7. Creating a HISTOGRAM for a categorical variable and shrink it

```python
sns.histplot(x='Payment',data=mart,shrink=0.9)
```

#### 8. Using different stats or calculation or aggregation in a HIST PLOT

```python
sns.histplot(x='Total',data=mart,stat='percent',kde=True)

#stat can take values ['count', 'frequency', 'density', 'probability', 'proportion', 'percent']
```

#### 9. Creating a bivariate HISTOGRAM using HISTPLOT

```python
sns.histplot(x='Total',y='gross income',data=mart)
```

## RUG PLOT

Plot marginal distributions by drawing ticks along the x and y axes

This function is inteded to complement other plots showing the location of individual observations in an unobstrusive way

```python
mart.head()
```

#### 1. Creating a basic RUG Plot using one numeric variable

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x='Unit price')
```

#### 2. Creating a basic RUG Plot using two numeric variable

```python
plt.figure(figsize=(15,5))
sns.rugplot(x='Unit price',y='gross income',data=mart)
```

#### 3. Grouping the RUG Plot ticks by categories/categorical variable using Hue

```python
plt.figure(figsize=(15,5))
sns.rugplot(x='Unit price',y='gross income',data=mart,hue='Gender')
```

#### 4. Adjusting the height of RUGs/Ticks in a RUG Plot

```python
plt.figure(figsize=(15,5))
sns.rugplot(x='Unit price',y='gross income',data=mart,hue='Gender',height=0.08)
```

#### 5. Combining a RUG Plot with a KDE Plot

```python
plt.figure(figsize=(15,5))
sns.kdeplot(x='Unit price',y='gross income',data=mart,hue='Gender',fill=True)
sns.rugplot(x='Unit price',y='gross income',data=mart,hue='Gender',height=0.08)
```

#### 6. Combining a RUG Plot with a Scatter Plot

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x='Unit price',y='gross income',data=mart,hue='Gender')
sns.rugplot(x='Unit price',y='gross income',data=mart,hue='Gender',height=0.08)
```

#### 7. Showing Rugs/Ticks out side of the Plot

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x='Unit price',y='gross income',data=mart,hue='Gender')
sns.rugplot(x='Unit price',y='gross income',data=mart,hue='Gender',height=-0.01,clip_on=False)
```

## JOINT PLOT

Draw a plot of two variables with bivariate and univariate graphs

The plot consists of one relational plot and two marginal distribution plots(on x and y axes)

```python
iris = sns.load_dataset('iris')
iris.head()
```

#### 1. Creating a basic JOINT PLOT

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length')
```

#### 2. Changing the Plot kind in a JOINT PLOT to SCATTER, KDE, HIST, HEX, REG or RESID plot

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',kind='kde',fill=True,color='orange')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',kind='hist',color='green')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',kind='hex',color='violet')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',kind='reg',color='yellow')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',kind='resid',color='brown')
```

#### 3. Categorization in a JOINT PLOT basis on the value of a categorical variable using Hue arguement

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

#### 4. Changing the format of individual plots in a JOINT PLOT|Changing the format/style of Relational and Marginal Distribution plot separatly in a JOINT PLOT

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              joint_kws=dict(marker='*',color='black'),
              marginal_kws=dict(color='green',kde=True))
```

#### 5. Adjusting Height, Ratio and Space in a JOINT PLOT

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              height=8,
              ratio=3,
              space=0.5)
```

#### 6. Adding Y axis to the Marginal Distribution Plot in a JOINT PLOT| Showing/Hiding the Marginal Ticks of a Distribution Plot in a JOINT PLOT

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',
              height=8,
              ratio=3,
              space=0.5,
              marginal_ticks=True,
             color='purple')
```

#### 7. Creating KDE and RUG Plot on top of the JOINT PLOT

```python
l = sns.jointplot(data=iris,x='sepal_length',y='petal_length',joint_kws=dict(color='orange'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot)
```
